#Salary datset regression

```python
#step1 import pandas
import pandas as pd
```

```python
#step2 import data set
house =pd.read_csv('https://github.com/YBI-Foundation/Dataset/raw/main/Boston.csv')
```

```python
#step3know your datset
house.head()
```

```python
house.info()
```

```python
house.describe()
```

```python
house.columns
```

```python
#step4 Identify your target function and features .
x=house[['CRIM', 'ZN', 'INDUS', 'CHAS', 'NX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX',
       'PTRATIO', 'B', 'LSTAT']]
y=house['MEDV']
```

```python
x
y
```

```python
#slpit
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y, train_size=0.7,random_state=2529)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

#Select model

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

#prediction of our model


```python
y_pred = model.predict(x_test)

y_pred
```

```python
x_test
```

```python
from sklearn.metrics import mean_absolute_percentage_error
```

```python
mean_absolute_percentage_error(y_test,y_pred)
```
