```python
plt.figure(figsize=(15,5))
plt.bar(list(ipl['player_of_match'].value_counts()[0:10].keys()),ipl['player_of_match'].value_counts()[0:10],color='pink')
plt.show()
```

```python
ipl['result'].value_counts()
```

```python
ipl['toss_winner'].value_counts()
```

```python
batting_first=ipl[ipl['win_by_runs']!=0]
```

```python
batting_first.head()
```

```python
plt.figure(figsize=(5,5))
plt.hist(batting_first['win_by_runs'])
plt.title("Distribution of Runs")
plt.xlabel("runs")
plt.ylabel("")
plt.show
```

```python
batting_first['winner'].value_counts()
```

```python
from pandas.core.base import value_counts
plt.figure(figsize=(10,4))
plt.bar(list(batting_first['winner'].value_counts()[0:5].keys()),list(batting_first['winner'].value_counts()[0:5]),color=["red","yellow","green","blue","black"] )
plt.show()
```

```python
#Making a piechart
plt.figure(figsize=(7,7))
plt.pie(list(batting_first['winner'].value_counts()),labels=list(batting_first['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show
```

```python
#extracting those records in which teams win after batting second
batting_second=ipl[ipl['win_by_wickets']!=0]
```

```python
batting_second.head()
```

```python
#making a histogram for frequency of no wins w.r.t number of wickets
plt.figure(figsize=(7,7))
plt.hist(batting_second['win_by_wickets'],bins=30)
plt.xlabel('frequency')
plt.ylabel('no of wickets')
plt.show
```

```python
#finding out the frequency of nummber of wines after batting second
batting_second['winner'].value_counts()
```

```python
#making a bar plot for top 5 teams wins after they bats seconds
plt.figure(figsize=(15,7))
plt.bar(list(batting_second['winner'].value_counts()[0:5].keys()),list(batting_second['winner'].value_counts()[0:5]),color=['purple','blue','orange','yellow','pink'])
plt.show()
```

```python
#making a pie chart for disrribution of most wins after batting second
plt.figure(figsize=(7,7))
plt.pie(list(batting_second['winner'].value_counts()),labels=list(batting_second['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show
```
